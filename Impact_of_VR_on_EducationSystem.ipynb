```python
data = pd.read_csv('Virtual_Reality_in_Education_Impact.csv')


# Function to match a column element and return connected columns
def get_student_info(data, student_id):
    # Filter the DataFrame based on the student_id
    matched_row = data[data['Student_ID'] == student_id]
    
    if matched_row.empty:
        return f'No match found for Student ID {student_id}'
    
    return matched_row

# Take Student ID input from the user
student_id = input('Enter the Student ID to search: ')

result = get_student_info(data, student_id)
print(result)
```

```python
# 8.1 Importing Required Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import seaborn as sns

# 8.2 Preparing Data for Machine Learning
# Selecting relevant features
features = data[['Hours_of_VR_Usage_Per_Week', 'Engagement_Level', 'Instructor_VR_Proficiency']]
target = data['Improvement_in_Learning_Outcomes']

# Dropping rows with NaN values
features = features.dropna()
target = target[features.index]

# 8.3 One-Hot Encoding for Categorical Features
encoder = OneHotEncoder(sparse=False)
encoded_features = encoder.fit_transform(features)

# 8.4 Label Encoding for Target Variable if it's categorical
label_encoder = LabelEncoder()
encoded_target = label_encoder.fit_transform(target)

# 8.5 Splitting the Data
X_train, X_test, y_train, y_test = train_test_split(encoded_features, encoded_target, test_size=0.2, random_state=42)

# 8.6 Training the Model
model = LinearRegression()
model.fit(X_train, y_train)

# 8.7 Making Predictions
y_pred = model.predict(X_test)

# 8.8 Evaluating the Model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')

# 8.9 Visualizing Actual vs Predicted
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='#28A745')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.title('Actual vs Predicted Improvement in Learning Outcomes', fontsize=16)
plt.xlabel('Actual Values', fontsize=14)
plt.ylabel('Predicted Values', fontsize=14)
plt.grid()
plt.show()

# 8.10 Visualizing Residuals
plt.figure(figsize=(10, 6))
sns.residplot(x=y_pred, y=y_test - y_pred, lowess=True, color='blue')
plt.title('Residuals vs Fitted', fontsize=16)
plt.xlabel('Fitted Values', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.axhline(0, color='red', linestyle='--')  # Add a horizontal line at 0 for reference
plt.grid()
plt.show()
```

```python
# 8.12 Evaluating the Model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Display metrics
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R Score: {r2:.2f}')
```

```python
# 8.13 Visualizing Actual vs Predicted Values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', marker='o', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2)  # Diagonal line
plt.title('Actual vs Predicted Values', fontsize=16)
plt.xlabel('Actual Values', fontsize=14)
plt.ylabel('Predicted Values', fontsize=14)
plt.grid()
plt.show()
```
